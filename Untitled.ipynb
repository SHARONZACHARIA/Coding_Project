```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


column_name = 'Annual Salary' 


# Read data from the file
data = pd.read_csv('Dataset/data7.csv',names=[column_name])  # Replace 'your_data_file.csv' with your actual data file name

# Extract the column containing annual salaries
salaries = data['Annual Salary']

# Create a histogram to represent the probability density function (PDF)
plt.figure(figsize=(8, 6))
plt.hist(salaries, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Calculate mean annual salary (W̃)
mean_salary = np.mean(salaries)

# Calculate another required value (X) - Replace this with your specific calculation
# For example, if X represents the 75th percentile of salaries:
X_value = np.percentile(salaries, 75)

# Plot mean salary and X on the graph
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=1, label=f'Mean Salary ($W̃$): {mean_salary:.2f}')
plt.axvline(X_value, color='green', linestyle='dashed', linewidth=1, label=f'X: {X_value:.2f}')

# Set labels, title, and legend
plt.xlabel('Annual Salary')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Annual Salaries')
plt.legend()

# Show values of W̃ and X on the graph
plt.text(0.5, 0.95, f'Mean Salary ($W̃$): {mean_salary:.2f}', ha='center', va='center', transform=plt.gca().transAxes, color='red')
plt.text(0.5, 0.90, f'X: {X_value:.2f}', ha='center', va='center', transform=plt.gca().transAxes, color='green')

# Show the plot
plt.grid(True)
plt.show()
```
